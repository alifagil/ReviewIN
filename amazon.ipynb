```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score

# 1. Load Dataset
df = pd.read_csv('fake reviews dataset.csv')

# 2. Label Encoding (OR: 0, CG: 1)
df['label'] = df['label'].map({'OR': 0, 'CG': 1})

# 3. Preprocessing: Pastikan text adalah string dan hapus baris kosong
df['text_'] = df['text_'].astype(str).str.lower()
df = df.dropna(subset=['text_'])

# 4. Split Data
X_train, X_test, y_train, y_test = train_test_split(
    df['text_'], df['label'], test_size=0.2, random_state=42
)

# 5. Feature Extraction (TF-IDF)
# Ditambah ngram_range biar model belajar gabungan 2 kata (misal: "very good")
vectorizer = TfidfVectorizer(max_features=5000, ngram_range=(1,2))
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)

# 6. Training Model
model = LogisticRegression(max_iter=1000) # Ditambah max_iter biar konvergen
model.fit(X_train_tfidf, y_train)

# 7. Evaluasi
y_pred = model.predict(X_test_tfidf)
print(f"Accuracy: {accuracy_score(y_test, y_pred):.2f}")
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
import joblib

# Simpan model dan penterjemah teksnya (vectorizer)
joblib.dump(model, 'model_fraud.pkl')
joblib.dump(vectorizer, 'tfidf_vectorizer.pkl')

print("File model_fraud.pkl dan tfidf_vectorizer.pkl berhasil dibuat!")
```
